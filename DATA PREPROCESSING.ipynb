Tuy nhiên, trước khi phân chia, chúng em cần xử lý 2 vấn đề
- Các điểm dữ liệu bị sai tên quận --> dùng replace()


```python
df['neighbourhood group'] = df['neighbourhood group'].str.replace(r'brookln', 'Brooklyn')
df['neighbourhood group'] = df['neighbourhood group'].str.replace(r'manhatan', 'Manhattan')
```

- Các điểm dữ liệu NaN --> dùng Nearest neighbors

```python
#mã hoá các tên quận thành các nhãn
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
df['neighbourhood group']= label_encoder.fit_transform(df['neighbourhood group']) 
df['neighbourhood group'].unique() 
```

```python
#chuyển nhãn 5 thành nhãn NaN
df['neighbourhood group'].replace([5], np.nan, inplace = True)
```

```python
#import KNN
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=7)
df_1 = df[['long','lat','neighbourhood group']]
```

```python
#fill NaN
fill_missing_data = imputer.fit_transform(df_1)
fill_data = pd.DataFrame(fill_missing_data, columns=df_1.columns)
fill_data.loc[df_1[df_1.isna().any(axis=1)].index, :]
```

```python
df[['long','lat','neighbourhood group']]=fill_data
```

```python
df=df.dropna(subset=['long','lat'])
```

```python
#check null
df.isnull().sum()
```

```python
#chuyển nhãn số thành tên quận như ban đầu
df['neighbourhood group'] = df['neighbourhood group'].astype(str).str.replace(r'2.0', 'Manhattan')
df['neighbourhood group'] = df['neighbourhood group'].astype(str).str.replace(r'1.0', 'Brooklyn')
df['neighbourhood group'] = df['neighbourhood group'].astype(str).str.replace(r'3.0', 'Queens')
df['neighbourhood group'] = df['neighbourhood group'].astype(str).str.replace(r'0.0', 'Bronx')
df['neighbourhood group'] = df['neighbourhood group'].astype(str).str.replace(r'4.0', 'Staten Island')
```

```python
count = pd.DataFrame(df[['neighbourhood group']].value_counts())
count
```
